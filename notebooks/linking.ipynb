```python
import pandas as pd
```

```python
link_data = pd.read_csv('ml-25m/links.csv')
main_data = pd.read_csv('movies.csv')
imdb_data = pd.read_csv('imdb_movies.csv')
```

```python
# rename columns
imdb_data = imdb_data.rename(columns={'names': 'title'})
```

```python
import re
def remove_text_in_brackets(input_string):
    # Use a regular expression to match text within brackets and remove it
    result = re.sub(r'\([^)]*\)', '', input_string)
    result = result.strip() 
    return result
```

```python
merge_movies = pd.merge(main_data, link_data, on='movieId', how='left')
main_data['year'] = main_data['title'].str.extract('.*\((.*)\).*', expand=True)
main_data
```

```python
merge_movies['title'] = merge_movies['title'].apply(remove_text_in_brackets)
```

```python
merge_movies.to_csv('merge_movies.csv', index=False)
```
