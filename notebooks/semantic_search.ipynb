```python
import numpy as np
import torch
from transformers import AutoTokenizer, AutoModel

# Load pre-trained transformer model and tokenizer
model_name = "distilbert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name)

# Example movie dataset
movies = [
    "Star Wars: A New Hope is a classic space opera film directed by George Lucas.",
    "Happy new year is a Bollywood movie directed by Farah Khan.",
    "The Lord of the Rings is a fantasy film directed by Peter Jackson.",
    "The Godfather is a crime film directed by Francis Ford Coppola.",
    "Inception is a science fiction action film directed by Christopher Nolan.",
    "The Shawshank Redemption is a drama film directed by Frank Darabont.",
    "Pulp Fiction is a black comedy crime film directed by Quentin Tarantino."
]

# Preprocess and encode the movie dataset
encoded_movies = []
for movie in movies:
    encoded_input = tokenizer(movie, padding=True, truncation=True, return_tensors="pt")
    with torch.no_grad():
        output = model(**encoded_input)
    embeddings = output.last_hidden_state.mean(dim=1).squeeze().numpy()  # Mean pooling
    encoded_movies.append(embeddings)

# Function to perform semantic search
def semantic_search(query, encoded_docs, top_n=3):
    encoded_query = tokenizer(query, padding=True, truncation=True, return_tensors="pt")
    with torch.no_grad():
        output = model(**encoded_query)
    query_embedding = output.last_hidden_state.mean(dim=1).squeeze().numpy()  # Mean pooling

    # Calculate cosine similarity between query and documents
    similarities = []
    for doc_embedding in encoded_docs:
        cosine_sim = np.dot(query_embedding, doc_embedding) / (np.linalg.norm(query_embedding) * np.linalg.norm(doc_embedding))
        similarities.append(cosine_sim)

    # Get indices of top similar documents
    top_indices = np.argsort(similarities)[::-1][:top_n]
    top_results = [(movies[i], similarities[i]) for i in top_indices]
    return top_results

# Example query
query = "What are some dark movies?"

# Perform semantic search
results = semantic_search(query, encoded_movies)
for movie, similarity in results:
    print(f"Movie: {movie}, Similarity: {similarity}")
```
